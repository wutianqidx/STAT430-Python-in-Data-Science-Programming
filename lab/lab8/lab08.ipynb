```python
import pandas as pd
import numpy as np

name = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
         'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
indices = [2, 3, 8, 9, 11, 15, 18, 31, 37, 39, 40, 43, 50, 57]
data_list, l = [], []

with open('cleveland.data',encoding="utf8", errors='ignore') as f:
    for line in f:
        for word in line.split():
            l.append(word)
            if word == 'name' and len(l) == 76:
                l_14 = [float(l[i]) for i in indices]
                data_list.append(l_14)
                l = []
                
data = pd.DataFrame(data_list, columns = name)
data = data.astype({'num':'int64'})
data.replace(-9.0, np.nan, inplace=True)
```
