## 1. DataFrame Filtering

In this lab, we will again be working with the cleaned version of the Ramen Ratings dataset. There are 5 variables:

  - `'Brand'`: the ramen brand/company
  - `'Variety'`: the name of the ramen on the label
  - `'Style'`: the style of ramen (in term of packaging)
  - `'Country'`: the country producing the ramen
  - `'Stars'`: the rating

- This dataset is available at: https://stat430.hknguyen.org/files/datasets/clean-ramen.csv


Load the Ramen Ratings dataset to Python and store it in a variable called `ramen`.

```python
# your code here
import pandas as pd
ramen = pd.read_csv('https://stat430.hknguyen.org/files/datasets/clean-ramen.csv')
ramen.head()
```

### Exercise 1.1

Filter only the observations where the ramen is made in Japan and in Cup, OR the ramen is made in Thailand and in Pack.

```python
# your code here
filter_1 = ((ramen.Country == 'Japan') & (ramen.Style == 'Cup')) \
            | ((ramen.Country == 'Thailand') & (ramen.Style == 'Pack'))

ramen[filter_1]
```

### Exercise 1.2

Filter only the observations where the ramen had **at least 4 stars** and it is made in `USA` or `United States`.

```python
# your code here
filter_2 = (ramen.Stars >=4) & (ramen.Country.isin(['USA', 'United States']))
ramen[filter_2]
```

### Exercise 1.3

Filter only the `Brand` and `Variety` attributes (columns) of the observations where the ramen had at least 4 stars and it is made in `Vietnam`.

```python
# your code here
filter_3 = (ramen.Stars >= 4) & (ramen.Country == 'Vietnam')
ramen.loc[filter_3, ['Brand','Variety']]
```

### Exercise 1.4

Filter only the `Variety` and `Stars` attributes of the observations where the ramen had **at most 2 stars** and the brand is `Nissin`.

```python
# your code here
filter_4 = (ramen.Stars <= 2) & (ramen.Brand == 'Nissin')
ramen.loc[filter_4, ['Variety','Stars']]
```

### Exercise 1.5

Filter only the `Variety` and `Stars` attributes of the observations where the ramen had **at most 2 stars** and it is made in `United States` or `USA`.

```python
# your code here
filter_5 = (ramen.Stars <= 2) & (ramen.Country.isin(['United States', 'USA']))
ramen.loc[filter_5, ['Variety','Stars']]
```

### Exercise 1.6 (bonus)

Filter and sort all ramen produced in the US by descending `Stars` rating and alphabetical-ordered `Brand`.

```python
# your code here
filter_6 = ramen.Country.isin(['United States', 'USA'])
ramen[filter_6].sort_values(by = ['Stars', 'Brand'], ascending = [False, True])
```

## 2. Computing Basic Statistics on Subset of a DataFrame



### Exercise 2.1

Compute the mean stars rating of Nissin ramen (of brand Nissin).

```python
# your code here
ramen.loc[ramen.Brand == 'Nissin', 'Stars'].mean()
```

### Exercise 2.2

Compute the mean stars rating of all ramen made in the US (that is `USA` and `United States`).

```python
# your code here
ramen.loc[ramen.Country.isin(['USA', 'United States']), 'Stars'].mean()
```

### Exercise 2.3

Find the maximum stars rating of a Vifon ramen (of brand Vifon).

```python
# your code here
ramen.loc[ramen.Brand == 'Vifon', 'Stars'].max()
```

### Exercise 2.4

Now, find the observation corresponding to the value found in Ex 3.3. That is, find and print the whole row of the Vifon ramen that has stars rating of 5.0.

```python
# your code here
ramen.loc[(ramen.Brand == 'Vifon') & (ramen.Stars == 5)]
```

### Exercise 2.5

Find the observation (meaning the whole row) of the a Nongshim ramen with the lowest stars rating.
