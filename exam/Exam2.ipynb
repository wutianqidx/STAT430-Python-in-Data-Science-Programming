```python
'''
For each row with a missing value in 'Stars',
Find all rows in the ramen DataFrame with the same 'Brand' and 'Country', compute the average 'Stars'.
Use this value to fill in the missing value for 'Stars'.
'''
import pandas as pd
ramen = pd.read_csv('ramen-ratings.csv', na_values = 'Unrated')
ramen = ramen.drop('Review #', axis = 1)
ramen.loc[ramen.Country == 'United States', 'Country'] = 'USA'
brand = ramen.loc[ramen.Stars.isnull(), 'Brand'].to_list()
country = ramen.loc[ramen.Stars.isnull(), 'Country'].to_list()
def find_mean(ramen, brand, country):
    mean = []
    for i in range(len(brand)):
        condition = (ramen.Brand == brand[i]) & (ramen.Country == country[i])
        mean.append(ramen.loc[condition, 'Stars'].mean())
    return mean
ramen.loc[ramen.Stars.isnull(), 'Stars'] = find_mean(ramen, brand, country)
```

```python
'''If you look at cleveland, you will notice that the only columns with missing values are: 'ca' and 'thal'.
For each row with missing data,
Using 'sex' (sex) and 'num' (diagnosis) of this row, 
find all the rows in the cleveland DataFrame with matching sex and diagnosis, 
then find the most common value for the missing variable.
Use that value to fill in the missing value.
'''
import pandas as pd
import numpy as np
name = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
         'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']

cleveland = pd.read_csv('processed.cleveland.data', na_values = '?', names = name)
miss_thal = cleveland.loc[cleveland.thal.isnull()].index
miss_ca = cleveland.loc[cleveland.ca.isnull()].index
def find_mode(cleveland, miss_val, name):
    for i in miss_val:
        condition = (cleveland.sex == cleveland.loc[i, 'sex']) & (cleveland.num == cleveland.loc[i, 'num'])
        cleveland.loc[i, name] = cleveland.loc[condition, name].mode()[0]
    return

find_mode(cleveland, miss_thal, 'thal')
find_mode(cleveland, miss_ca, 'ca')
```

## 7.7

```python
'''
Depending on 'slope', the values of 'thalach' are classified as 'high' or 'low':
Upsloping: >= 160 is high, low otherwise.
Flat: >= 140 is high, low otherwise.
Downsloping: >= 145 high, low otherwise.
Create a new variable (column) called 'thalach_level' which has two possible values: 
'high' or 'low' depending on 'slope' and 'thalach'.
'''
import pandas as pd
import numpy as np
name = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
         'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']

cleveland = pd.read_csv('processed.cleveland.data', na_values = '?', names = name)
def modify(row):
    val = 'low'
    if row.cp == 1.0 and row.thalach >= 160:
        val = 'high'
    if row.cp == 2.0 and row.thalach >= 140:
        val = 'high'
    if row.cp == 3.0 and row.thalach >= 145:
        val = 'high'
    return val
cleveland['thalach_level'] = cleveland.apply(modify, axis = 1)
cleveland.head()
```
